```python
import numpy as np
import healpy as hp
from matplotlib import pyplot as plt
import astropy.units as u
```

```python
ra = 360.*np.random.random(100000)
dec = (180./np.pi)*np.arcsin(1.-np.random.random(100000)*2)
```

```python
area = hp.nside2pixarea(nside=1, degrees=True)
print(f'Each pixel has {area} degree^2')
```

```python
loc = hp.ang2pix(nside=1, lonlat=True, theta=ra, phi=dec)
pixel, counts = np.unique(loc, return_counts=True)
list(zip(pixel, counts))
```

## Each pixel roughly has same number of counts, this means they have same area.

```python
in_pixel_2 = loc == 1
in_pixel_5 = loc == 4
in_pixel_8 = loc == 7
```

```python
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(121, projection='aitoff')
ax.scatter(ra*u.deg.to(u.rad)-np.pi, dec*u.deg.to(u.rad), marker='o', color='b', s=0.7, alpha=0.5)
ax.scatter(ra[in_pixel_2]*u.deg.to(u.rad)-np.pi, dec[in_pixel_2]*u.deg.to(u.rad), marker='o', color='r', s=0.7, alpha=0.5)
xlab = [f'{str(i)}h' for i in np.arange(14, 36, 2)%24]
ax.set_xticklabels(xlab, weight=100)
ax.grid(color='k', linestyle='--', linewidth=0.5)
```
