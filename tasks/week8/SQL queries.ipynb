```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
targets = pd.read_csv('sql_result.csv', sep=',')
```

```python
plt.plot(targets['ra'], targets['dec'], 'bo')
plt.xlabel('ra')
plt.ylabel('dec')
plt.show()
```

```python
large_g = targets['modelMag_g'] <= 17
plt.scatter(targets['ra'], targets['dec'], s=4, label='low g')
plt.scatter(targets['ra'][large_g], targets['dec'][large_g], s=16, label='high g')
plt.xlabel('ra')
plt.ylabel('dec')
plt.legend()
plt.show()
```
